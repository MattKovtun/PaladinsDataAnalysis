```python
def select_macthes(hero_name, win=True):
    ddf = df[(df['hero'] == hero_name) & (df['winner'] == win)]
    match_ids = ddf['matchid']
    
    return df[(df['matchid'].isin(match_ids)) 
              & (df['hero'] != hero_name) 
              & (df['winner'] == (not win))]
```

```python
won_matches = select_macthes(hero).groupby('hero').count()
lost_matches = select_macthes(hero, win=False).groupby('hero').count()
1 - (won_matches / (won_matches + lost_matches))
```

```python
data = df[(df['tier'] >= 11) & (df['tier'] <= 20)]['matchid'].unique()
```

```python
data = df[df['matchid'].isin(data)]
```

```python
heroes = data['hero'].unique()
heroes.sort()
heroes = {hero: i for i, hero in enumerate(heroes)}
```

```python
dd = data.groupby(['matchid', 'winner'])
```

```python
dd.first()
```

```python
X = []
y = []
```

```python
maps = df['map'].unique()
maps = {m: i for i, m in enumerate(maps)}
```

```python
for i in dd:
    winner = i[0][1]
    if winner:
        y.append(1)
    else:
        y.append(0)
        
    tmp = [0] * len(heroes)
    for ix, j in enumerate(i[1]['hero']): 
        tmp[heroes[j]] = i[1]['tier'].tolist()[ix]
        
#     tmp.append(maps[i[1]['map'].tolist()[0]])
    X.append(tmp)
    
    
```

```python
len(X), len(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score


train_x, test_x, train_y, test_y = train_test_split(X, y)
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(max_depth=6)
xgb.fit(np.array(train_x), np.array(train_y))
y_pred = xgb.predict(test_x)
accuracy_score(test_y, y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(np.array(train_x), np.array(train_y))
y_pred = lr.predict(test_x)
accuracy_score(test_y, y_pred)
```
